```python
Y_train=train_df["subject"]
X_train=train_df.drop("subject",axis=1)





unique_subjects=train_df["subject"].drop_duplicates()
unique_subjects.tail(20)
# unique_subjects.info()


```

# For Crossvalidation

Creating sample space of all the subjects. The data is to be choosen at _random_ and the data samples that is collected for every subject are the same


```python
X_train.head(1)
```

```python
# Logistic Regression

logreg = LogisticRegression()

logreg.fit(X_train, Y_train)

#Y_pred = logreg.predict(X_test)

logreg.score(X_train, Y_train)
```

```python
# Support Vector Machines

svc = SVC()

svc.fit(X_train, Y_train)

# Y_pred = svc.predict(X_test)

svc.score(X_train, Y_train)
```

```python
# Random Forests

random_forest = RandomForestClassifier(n_estimators=100)

random_forest.fit(X_train, Y_train)

# Y_pred = random_forest.predict(X_test)

random_forest.score(X_train, Y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)

knn.fit(X_train, Y_train)

# Y_pred = knn.predict(X_test)

knn.score(X_train, Y_train)
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()

gaussian.fit(X_train, Y_train)

# Y_pred = gaussian.predict(X_test)

gaussian.score(X_train, Y_train)
```
